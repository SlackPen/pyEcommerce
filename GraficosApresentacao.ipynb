# Gr√°ficos

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dados das empresas e classifica√ß√µes
dados = {
    'Empresa': ['Vulcabras', 'GPA', 'Serena', 'PRINER'],
    'a-Alto': [20, 35, 30, 35],
    'b-M√©dio': [25, 32, 34, 20],
    'c-Baixo': [30, 25, 27, 25],
    'x-Com certeza Sim': [35, 22, 30, 35],
    'z-Com certeza N√£o': [25, 27, 31, 24]
}

# Cria√ß√£o do DataFrame
df = pd.DataFrame(dados)

# Definir as categorias de classifica√ß√£o
categorias = ['a-Alto', 'b-M√©dio', 'c-Baixo', 'x-Com certeza Sim', 'z-Com certeza N√£o']

# Configurar o tamanho do gr√°fico
plt.figure(figsize=(10, 6))

# Definir a posi√ß√£o das barras no eixo X
import numpy as np
bar_width = 0.15
r1 = np.arange(len(df['Empresa']))
for i, categoria in enumerate(categorias):
    plt.bar(r1 + i * bar_width, df[categoria], width=bar_width, label=categoria)

# Adicionar labels e t√≠tulo
plt.xlabel('Empresas')
plt.ylabel('Quantidade de Textos')
plt.title('Classifica√ß√£o de Textos por Empresa')
plt.xticks([r + bar_width * 2 for r in range(len(df['Empresa']))], df['Empresa'])
plt.legend()

# Exibir o gr√°fico
plt.tight_layout()
plt.show()
```
